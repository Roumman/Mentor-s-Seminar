```python
TRINO_HOST = "localhost"
TRINO_PORT = 8080
TRINO_USER = "guest"
TRINO_HTTP_SCHEME = "http"
```

```python
import trino
import pandas as pd

conn = trino.dbapi.connect(
    host=TRINO_HOST,
    port=TRINO_PORT,
    user=TRINO_USER,
    http_scheme=TRINO_HTTP_SCHEME,
)

cur = conn.cursor()
cur.execute("SELECT 1 AS ok")
cur.fetchall()
```

```python
cur.execute("SHOW CATALOGS")
catalogs = [r[0] for r in cur.fetchall()]
catalogs
```

```python
cur.execute("SHOW SCHEMAS FROM postgresql")
schemas = [r[0] for r in cur.fetchall()]
schemas[:50]
```

```python
schema = "public"

cur.execute(f"SHOW TABLES FROM postgresql.{schema}")
tables = [r[0] for r in cur.fetchall()]
tables
```

```python
query = """
SELECT id, name, age
FROM postgresql.public.users
ORDER BY id
LIMIT 10
"""

cur.execute(query)
rows = cur.fetchall()
cols = [d[0] for d in cur.description]

df = pd.DataFrame(rows, columns=cols)
df
```

```python
query = """
SELECT count(*) AS cnt
FROM postgresql.public.users
"""

cur.execute(query)
rows = cur.fetchall()
cols = [d[0] for d in cur.description]

pd.DataFrame(rows, columns=cols)
```
